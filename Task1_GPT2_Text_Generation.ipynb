```python

# TEXT GENERATION WITH GPT-2
# PRODIGY INFOTECH TASK-01

!pip install transformers datasets torch accelerate -q

# Import libraries
from transformers import (
    GPT2LMHeadModel,
    GPT2Tokenizer,
    Trainer,
    TrainingArguments,
    pipeline,
    DataCollatorForLanguageModeling
)

from datasets import Dataset

# STEP 1: CREATE DATASET

texts = [
         "Generative AI creates brand-new content from scratch.",
         "Large language models are trained on vast amounts of text.",
         "Prompt engineering shapes how a model responds.",
         "Hallucination is when AI confidently states false information.",
         "Deepfakes use generative AI to create fake videos of real people.",
         "Fine-tuning adapts a general model to a specific task.",
         "Diffusion models generate images by learning to remove noise.",
         "RLHF trains AI to align with human preferences.",
         "Generative AI is revolutionizing how software is written.",
         "Agentic AI can autonomously complete multi-step tasks."
]

dataset = Dataset.from_dict({"text": texts})

print("Dataset Created Successfully!")

# STEP 2: LOAD GPT-2 MODEL

model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Fix padding token
tokenizer.pad_token = tokenizer.eos_token

print("GPT-2 Model Loaded!")

# STEP 3: TOKENIZATION

def tokenize_function(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=64
    )

tokenized_dataset = dataset.map(tokenize_function, batched=True)

print("Tokenization Completed!")

# STEP 4: DATA COLLATOR

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

# STEP 5: TRAINING ARGUMENTS

training_args = TrainingArguments(
    output_dir="./gpt2-output",
    num_train_epochs=3,
    per_device_train_batch_size=1,
    save_steps=10,
    logging_steps=5
)

# STEP 6: TRAIN MODEL

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)

print("Training Started...\n")

trainer.train()

print("\nTraining Completed!")

# STEP 7: SAVE MODEL

model.save_pretrained("./gpt2-model")
tokenizer.save_pretrained("./gpt2-model")

print("Model Saved Successfully!")

# STEP 8: GENERATE TEXT

generator = pipeline(
    "text-generation",
    model="./gpt2-model",
    tokenizer=tokenizer
)

prompt = "Generative AI"
result = generator(
    prompt,
    max_length=50,
    num_return_sequences=1,
    temperature=0.7
)


print("\nGenerated Text:\n")

generated_text = result[0]['generated_text']

sentences = generated_text.split(".")

for sentence in sentences:
    if sentence.strip() != "":
        print(sentence.strip() + ".")
```
